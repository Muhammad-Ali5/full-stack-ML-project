# Student Performance EDA

This notebook explores the student performance dataset step by step. It covers loading the data, checking its structure, understanding the target variable, and building initial insights for the modeling stage.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
sns.set_theme(style='whitegrid')

DATA_PATH = os.path.join('src', 'notebook', 'data', 'student_performance_data.csv')
df = pd.read_csv(DATA_PATH)

print('Shape of the dataset:', df.shape)
df.head()
```

## Dataset Overview

This section shows the dataset shape and the first rows so the structure is clear before deeper analysis.

```python
df.info()
print('\nColumns:', df.columns.tolist())
print('\nMissing values:\n', df.isnull().sum())
```

## Target Distribution

The target column is the final grade, which is a categorical label. Understanding its class balance helps decide how to evaluate the model.

```python
target_counts = df['grade'].value_counts()
print(target_counts)

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='grade', order=target_counts.index)
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Numerical Feature Relationships

This section explores how key numerical variables relate to the target score and grade outcomes.

```python
numeric_cols = ['study_hours_per_day','attendance_percentage','assignment_score','midterm_score','final_exam_score','participation_score','sleep_hours','overall_score']

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x='grade', y=col)
    plt.title(f'{col} by Grade')
    plt.xlabel('Grade')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

correlation_matrix = df[numeric_cols + ['overall_score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
